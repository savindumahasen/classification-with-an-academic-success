## Model Selection and Training

```python
## splitting the training dataset into the training and validation set
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(training_set_pd,training_set_label, random_state=42, test_size=0.2)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
## check the length of X_train and y_train part
```

```python
if (len(X_train) == len(y_train)):
    print("Length is equal")
else:
    print("Length is not equal")
```

```python
## Random forest model
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier(random_state=42)
rf_clf.fit(X_train,y_train)
```

```python
## random forest classifier predictions
```

```python
rf_clf_predictions =rf_clf.predict(X_test)
rf_clf_predictions
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
## random forest classifier model validation accuracy (accuracy score)
```

```python
rf_validation_accuracy = accuracy_score(y_test, rf_clf_predictions)
print(f"\nValidation Accuracy: {rf_validation_accuracy:.4f}")
```

```python
## random forest model mean squared error
```

```python
rf_mse= mean_squared_error(y_test,rf_clf_predictions)
rf_mse
```

```python
## random forest model root mean squared error
```

```python
rf_rmse = np.sqrt(rf_mse)
rf_rmse
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_clf =DecisionTreeClassifier(random_state=42)
dt_clf.fit(X_train,y_train)
```

```python
## decision tree classifier predictions
```

```python
dt_clf_predictions =dt_clf.predict(X_test)
dt_clf_predictions
```

```python
## decision tree classifier model validation accuracy (accuracy score)
```

```python
dt_validation_accuracy = accuracy_score(y_test, dt_clf_predictions)
print(f"\nValidation Accuracy: {dt_validation_accuracy:.4f}")
```

```python
## decision tree classifier model mean squared error
```

```python
dt_mse =  mean_squared_error(y_test,dt_clf_predictions)
dt_mse
```

```python
## decision tree classifier model root mean squared error
```

```python
dt_rmse =np.sqrt(dt_mse)
dt_rmse
```

```python
from sklearn.linear_model import LogisticRegression

bl_clf = LogisticRegression(random_state=42, max_iter=1000, tol=1e-05)
bl_clf.fit(X_train,y_train)
```

```python
## binary logistic regression model predictions
```

```python
bl_clf_predictions = bl_clf.predict(X_test)
bl_clf_predictions
```

```python
## binary logistic regression model validation score (accuracy score)
```

```python
bl_clf_validations = accuracy_score(y_test, bl_clf_predictions)
```

```python
bl_clf_validations
```

```python
## binary logistic regression model mean squared error
```
